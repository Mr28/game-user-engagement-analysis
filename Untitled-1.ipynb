```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from scipy.stats import pointbiserialr, chi2_contingency
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, f1_score, precision_score, recall_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv(filepath_or_buffer=r"technical_test_data(in).csv")
```

```python
df['serverdate_first'], df['serverdate_last'], df['clientdate_first'], df['clientdate_last'] = pd.to_datetime(df['serverdate_first']), pd.to_datetime(df['serverdate_last']), pd.to_datetime(df['clientdate_first']), pd.to_datetime(df['clientdate_last'])
```

```python
# drop RU5514 (session_playtime_duration==-1) and NaN country codes
df = df[(df['countrycode']!='RU5514') & (df['countrycode'].notna())]
```

```python
df = df.groupby(['playerid', 'clientdate_first']).agg(
    # clientdate_first = ('clientdate_first', 'min'),
    clientdate_last = ('clientdate_last', 'max'),
    serverdate_first = ('serverdate_first', 'min'),
    serverdate_last = ('serverdate_last', 'max'),
    countrycode = ('countrycode', lambda x: pd.Series.mode(x)[0]),
    playersessions_index = ('playersessions_index', lambda x: pd.Series.mode(x)[0]),
    sessionuptime_duration = ('sessionuptime_duration', 'max'),
    session_playtime_duration = ('session_playtime_duration', 'max'),
    population = ('population', lambda x: pd.Series.mode(x)[0])    
).reset_index().sort_values(by=['playerid', 'clientdate_first'])
```

```python
df.loc[(df['session_playtime_duration']!=0) & (df['sessionuptime_duration']==0), 'sessionuptime_duration'] = df['session_playtime_duration']
```

```python
df['time_since_joined'] = (df['clientdate_first'] - df.groupby('playerid')['clientdate_first'].transform('min')).dt.total_seconds()
```

```python
df['play_ratio'] = df['session_playtime_duration'] / df['sessionuptime_duration']
df.loc[(df['session_playtime_duration']==0) & (df['sessionuptime_duration']==0), 'play_ratio'] = 0.0

df['gap'] = df.groupby('playerid')['clientdate_first'].diff().dt.total_seconds().fillna(0)
# df['gap'] = (df['clientdate_first']-df.groupby('playerid')['clientdate_last'].shift(periods=1)).dt.total_seconds()
```

```python
for var in ['gap', 'play_ratio']:

    df[f'expanding_mean_{var}'] = df.groupby('playerid')[var].transform(lambda x: x.expanding().mean())
    df[f'expanding_std_{var}'] = df.groupby('playerid')[var].transform(lambda x: x.expanding().std().fillna(0))

    df[f'rolling_mean_{var}_3'] = df.groupby('playerid')[var].transform(lambda x: x.rolling(window=3, min_periods=1).mean())#.reset_index(level=0, drop=True)
    df[f'rolling_std_{var}_3'] = df.groupby('playerid')[var].transform(lambda x: x.rolling(window=3, min_periods=1).std().fillna(0))
```

```python
df['gap_between_sessions'] = (df['serverdate_first']-df.groupby('playerid')['serverdate_last'].shift(periods=1)).dt.total_seconds() / 3600

df['mean_gap_between_sessions'] = df.groupby('playerid')['gap_between_sessions'].transform('mean')
df['std_gap_between_sessions'] = df.groupby('playerid')['gap_between_sessions'].transform('std')

max_date = df['serverdate_last'].max()

df['last_session'] = (df.groupby('playerid')['serverdate_last'].shift(periods=-1).isna()) & \
    (((max_date-df['serverdate_last']).dt.total_seconds()/3600) > df['mean_gap_between_sessions'] + 2*df['std_gap_between_sessions'])

df.loc[(df.groupby('playerid')['serverdate_last'].shift(periods=-1).isna()) & (df['std_gap_between_sessions'].isna()), 'last_session'] = True


df.drop(columns=['gap_between_sessions', 'mean_gap_between_sessions', 'std_gap_between_sessions'], inplace=True)
```

```python
from scipy.stats import pointbiserialr, chi2_contingency

# Define numerical and categorical columns
numerical_cols = ['time_since_joined', 'play_ratio', 'gap', 'expanding_mean_gap', 'expanding_std_gap', 
                  'rolling_mean_gap_3', 'rolling_std_gap_3', 'expanding_mean_play_ratio', 'expanding_std_play_ratio', 
                  'rolling_mean_play_ratio_3', 'rolling_std_play_ratio_3', 'sessionuptime_duration', 
                  'session_playtime_duration', 'playersessions_index']

categorical_cols = ['countrycode', 'population']
```

```python
# Point-biserial correlation for numerical variables with 'last_session'
print("Point-biserial correlations with 'last_session':")
for col in numerical_cols:
    if col in df.columns:
        corr, p = pointbiserialr(df['last_session'].astype(int), df[col])
        print(f"{col}: correlation = {corr:.3f}, p-value = {p:.3f}")
    else:
        print(f"{col} not in df")
```

```python
# Chi-square test for categorical variables with 'last_session'
print("\nChi-square tests with 'last_session':")
for col in categorical_cols:
    if col in df.columns:
        contingency_table = pd.crosstab(df[col], df['last_session'])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"{col}: chi2 = {chi2:.3f}, p-value = {p:.3f}")
    else:
        print(f"{col} not in df")
```

```python
# Split data by players to avoid data leakage (since sessions are per player)
from sklearn.model_selection import train_test_split

unique_players = df['playerid'].unique()
print(f"Total unique players: {len(unique_players)}")

# Split players into train (60%), val (20%), test (20%)
train_players, temp_players = train_test_split(unique_players, test_size=0.4, random_state=42)
val_players, test_players = train_test_split(temp_players, test_size=0.5, random_state=42)

print(f"Train players: {len(train_players)}")
print(f"Val players: {len(val_players)}")
print(f"Test players: {len(test_players)}")

# Create datasets
df_train = df[df['playerid'].isin(train_players)].copy()
df_val = df[df['playerid'].isin(val_players)].copy()
df_test = df[df['playerid'].isin(test_players)].copy()

print(f"\nTrain set size: {len(df_train)}")
print(f"Val set size: {len(df_val)}")
print(f"Test set size: {len(df_test)}")

# Check class distribution in each set
print("\nTrain set class distribution:")
print(df_train['last_session'].value_counts(normalize=True))
print("\nVal set class distribution:")
print(df_val['last_session'].value_counts(normalize=True))
print("\nTest set class distribution:")
print(df_test['last_session'].value_counts(normalize=True))
```

```python
# Prepare features for modelling
selected_features = numerical_cols + categorical_cols
print("Selected features:", selected_features)

# Function to preprocess data
def preprocess_data(df, train_dummies=None):
    df = df[selected_features + ['last_session']].copy()
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
    df = df.fillna(0)  # Simple imputation
    if train_dummies is not None:
        # Align columns to training set
        for col in train_dummies:
            if col not in df.columns:
                df[col] = 0
        df = df[train_dummies + ['last_session']]
    return df

df_train_prep = preprocess_data(df_train)
train_dummies = [col for col in df_train_prep.columns if col != 'last_session']
df_val_prep = preprocess_data(df_val, train_dummies)
df_test_prep = preprocess_data(df_test, train_dummies)

# Features and target
feature_cols = train_dummies
X_train = df_train_prep[feature_cols]
y_train = df_train_prep['last_session']
X_val = df_val_prep[feature_cols]
y_val = df_val_prep['last_session']
X_test = df_test_prep[feature_cols]
y_test = df_test_prep['last_session']

print(f"Training features shape: {X_train.shape}")
print(f"Validation features shape: {X_val.shape}")
print(f"Test features shape: {X_test.shape}")
```

```python
# Baseline: Always predict majority class (False)
y_test_pred_baseline = [False] * len(y_test)
y_test_pred_proba_baseline = [1 - y_test.mean()] * len(y_test)  # Probability of False

print("\nBaseline (Always Predict False) on Validation Set:")
print(classification_report(y_test, y_test_pred_baseline))
print(f"AUC: {roc_auc_score(y_test, y_test_pred_proba_baseline):.3f}")
print(f"F1 Score: {f1_score(y_test, y_test_pred_baseline):.3f}")
```

```python
# Benchmark Model: Logistic Regression with class weights
lr_model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
lr_model.fit(X_train, y_train)

# Predictions on test set
y_test_pred_lr = lr_model.predict(X_test)
y_test_pred_proba_lr = lr_model.predict_proba(X_test)[:, 1]

print("Logistic Regression on Validation Set:")
print(classification_report(y_test, y_test_pred_lr))
print(f"AUC: {roc_auc_score(y_test, y_test_pred_proba_lr):.3f}")
print(f"F1 Score: {f1_score(y_test, y_test_pred_lr):.3f}")
```

```python
# Hyperparameter Tuning for Random Forest using GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define parameter grid
param_grid = {
    'n_estimators': [10, 30, 100],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced']
}

# Initialize Random Forest
rf = RandomForestClassifier(random_state=42)

# Grid Search with cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='roc_auc', n_jobs=-1, verbose=2)
grid_search.fit(X_val, y_val)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
```

```python
# Comparison Model: Random Forest

rf_model = RandomForestClassifier(random_state=28)
rf_model.set_params(**grid_search.best_params_)
rf_model.fit(X_train, y_train)

# Predictions on test set
y_test_pred_rf = rf_model.predict(X_test)
y_test_pred_proba_rf = rf_model.predict_proba(X_test)[:, 1]

print("\nRandom Forest on Validation Set:")
print(classification_report(y_test, y_test_pred_rf))
print(f"AUC: {roc_auc_score(y_test, y_test_pred_proba_rf):.3f}")
print(f"F1 Score: {f1_score(y_test, y_test_pred_rf):.3f}")

# Feature importance
rf_importance = pd.Series(rf_model.feature_importances_, index=feature_cols).sort_values(ascending=False)
print("\nTop 10 RF Feature Importances:")
print(rf_importance.head(10))
```

```python
# Chi-square test for categorical variables with 'last_session'
print("\nChi-square tests with 'last_session':")
for col in categorical_cols:
    if col in df.columns:
        contingency_table = pd.crosstab(df[col], df['last_session'])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        print(f"{col}: chi2 = {chi2:.3f}, p-value = {p:.3f}")
    else:
        print(f"{col} not in df")
```

# Modelling

```python
# Prepare features for modelling
selected_features = numerical_cols + categorical_cols
print("Selected features:", selected_features)

# Function to preprocess data
def preprocess_data(df, train_dummies=None):
    df = df[selected_features + ['last_session']].copy()
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
    df = df.fillna(0)  # Simple imputation
    if train_dummies is not None:
        # Align columns to training set
        for col in train_dummies:
            if col not in df.columns:
                df[col] = 0
        df = df[train_dummies + ['last_session']]
    return df

df_train_prep = preprocess_data(df_train)
train_dummies = [col for col in df_train_prep.columns if col != 'last_session']
df_val_prep = preprocess_data(df_val, train_dummies)
df_test_prep = preprocess_data(df_test, train_dummies)

# Features and target
feature_cols = train_dummies
X_train = df_train_prep[feature_cols]
y_train = df_train_prep['last_session']
X_val = df_val_prep[feature_cols]
y_val = df_val_prep['last_session']
X_test = df_test_prep[feature_cols]
y_test = df_test_prep['last_session']

print(f"Training features shape: {X_train.shape}")
print(f"Validation features shape: {X_val.shape}")
print(f"Test features shape: {X_test.shape}")
```

```python
# Benchmark Model: Logistic Regression with class weights
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, f1_score

lr_model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
lr_model.fit(X_train, y_train)

# Predictions on validation set
y_val_pred_lr = lr_model.predict(X_val)
y_val_pred_proba_lr = lr_model.predict_proba(X_val)[:, 1]

print("Logistic Regression on Validation Set:")
print(classification_report(y_val, y_val_pred_lr))
print(f"AUC: {roc_auc_score(y_val, y_val_pred_proba_lr):.3f}")
print(f"F1 Score: {f1_score(y_val, y_val_pred_lr):.3f}")
```

```python
# Comparison Model: Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions on validation set
y_val_pred_rf = rf_model.predict(X_val)
y_val_pred_proba_rf = rf_model.predict_proba(X_val)[:, 1]

print("\nRandom Forest on Validation Set:")
print(classification_report(y_val, y_val_pred_rf))
print(f"AUC: {roc_auc_score(y_val, y_val_pred_proba_rf):.3f}")
print(f"F1 Score: {f1_score(y_val, y_val_pred_rf):.3f}")

# Feature importance
rf_importance = pd.Series(rf_model.feature_importances_, index=feature_cols).sort_values(ascending=False)
print("\nTop 10 RF Feature Importances:")
print(rf_importance.head(10))
```

```python
# Baseline: Always predict majority class (False)
y_val_pred_baseline = [False] * len(y_val)
y_val_pred_proba_baseline = [1 - y_val.mean()] * len(y_val)  # Probability of False

print("\nBaseline (Always Predict False) on Validation Set:")
print(classification_report(y_val, y_val_pred_baseline))
print(f"AUC: {roc_auc_score(y_val, y_val_pred_proba_baseline):.3f}")
print(f"F1 Score: {f1_score(y_val, y_val_pred_baseline):.3f}")
```

```python
# Hyperparameter Tuning for Random Forest using GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'class_weight': ['balanced']
}

# Initialize Random Forest
rf = RandomForestClassifier(random_state=42)

# Grid Search with cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='roc_auc', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best AUC Score:", grid_search.best_score_)

# Evaluate best model on validation set
best_rf = grid_search.best_estimator_
y_val_pred_best = best_rf.predict(X_val)
y_val_pred_proba_best = best_rf.predict_proba(X_val)[:, 1]

print("\nBest RF on Validation Set:")
print(classification_report(y_val, y_val_pred_best))
print(f"AUC: {roc_auc_score(y_val, y_val_pred_proba_best):.3f}")
print(f"F1 Score: {f1_score(y_val, y_val_pred_best):.3f}")
```
